# **Text-Cleaning**

```
df.info()
```

```
df.isnull().sum()
```

```
df['Emotion'].unique()
```

```
e_number = {}
i = 0
for e in df['Emotion'].unique():
    e_number[e] = i
    i += 1
df['Emotion'] = df['Emotion'].map(e_number)
```

```
df
```

LowerCaseing

```
df['Comment'] = df['Comment'].apply(lambda x: x.lower())
df
```

Remove Punctuations

```
import string
def remove_punctuation(txt):
  return txt.translate(str.maketrans('','',string.punctuation))
df['Comment'].apply(remove_punctuation)
```

Remove Numbers

```
def remove_number(txt):
  new = ""
  for i in txt:
    if not i.isdigit():
      new += i
  return new
df['Comment'].apply(remove_number)
```

Remove URL

```
import re
def remove_url(txt):
  new = "https?://\S+|www\.\S+"

  remove = re.sub(new,'',str(txt))
  return remove.strip()
df['Comment'] = df['Comment'].apply(remove_url)
df['Comment']
```

Remove HTML Tags

```
def remove_tag(txt):
  new = r"<.*?>"

  remove = re.sub(new,'',str(txt))
  return remove.strip()
df['Comment'] = df['Comment'].apply(remove_tag)
df['Comment']
```

Remove Special Characters & Emojis

```
def remove_emojis(txt):
  new = ""
  for i in txt:
    if i.isascii():
      new += i
  return new
df['Comment'] = df['Comment'].apply(remove_emojis)
df['Comment']
```

Remove Stopwords

```
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('punkt')
nltk.download('stopwords')
```

```
stopwords = set(stopwords.words('english'))
stopwords
```

```
nltk.download('punkt_tab')
def remove_stopwords(txt):
  word = word_tokenize(txt)
  cleaned = []
  for i in word:
    if i not in stopwords:
      cleaned.append(i)
  return ' '.join(cleaned)
df['Comment'] = df['Comment'].apply(remove_stopwords)
df.loc[1]['Comment']
```

Spelling check

```
from textblob import TextBlob
def correct_spell(txt):
  return str(TextBlob(txt).correct())
```

# **Model spliting**

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df['Comment'],df['Emotion'], test_size=0.20, random_state=42)
```

```
x_train
```

```
x_test
```

```
df.shape
```

# **Text-Preprocessing/Feature Extraction/Vectorization**

**1.Bag of Words**

```
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
```

```
bow = CountVectorizer()

x_train_bow = bow.fit_transform(x_train)
x_test_bow = bow.transform(x_test)
```

```
x_train_bow
```

```
x_test_bow
```

**2.TF-TDF Vectorizer**

```
tfidf = TfidfVectorizer()

x_train_tfidf = tfidf.fit_transform(x_train)
x_test_tfidf = tfidf.transform(x_test)
```

```
x_train_tfidf
```

```
x_test_tfidf
```

# **Training a Model**

```
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score,classification_report
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

1. naive bayes for bag of words

```
nb_model = MultinomialNB()
nb_model.fit(x_train_bow,y_train)
```

```
nb_pred = nb_model.predict(x_test_bow)
print(accuracy_score(y_test,nb_pred))
print(classification_report(y_test,nb_pred))
```

2.naive bayes for TF-IDF Vectorizer

```
nb_model_tf = MultinomialNB()
nb_model_tf.fit(x_train_tfidf,y_train)
```

```
nb_pred_tf = nb_model_tf.predict(x_test_tfidf)
print(accuracy_score(y_test,nb_pred_tf))
print(classification_report(y_test,nb_pred_tf))
```

1.SVM for Bag of Words

```
svm_model = SVC(kernel='linear', C=1)
svm_model.fit(x_train_bow,y_train)
```

```
svm_pred = svm_model.predict(x_test_bow)
print(accuracy_score(y_test,svm_pred))
print(classification_report(y_test,svm_pred))
```

2. SVM for TF-IDF Vectorizer

```
svm_model_tf = SVC(kernel='linear', C=1)
svm_model_tf.fit(x_train_tfidf,y_train)
```

```
svm_pred_tf = svm_model_tf.predict(x_test_tfidf)
print(accuracy_score(y_test,svm_pred_tf))
print(classification_report(y_test,svm_pred_tf))
```

1.Logistic Regression for Bag of words

```
lg_model = LogisticRegression(max_iter=5000)
lg_model.fit(x_train_bow,y_train)
```

```
lg_pred = lg_model.predict(x_test_bow)
print(accuracy_score(y_test,lg_pred))
print(classification_report(y_test,lg_pred))
```

```
lg_model_tf = LogisticRegression(max_iter=5000)
lg_model_tf.fit(x_train_tfidf,y_train)
```

```
lg_pred_tf = lg_model_tf.predict(x_test_tfidf)
print(accuracy_score(y_test,lg_pred_tf))
print(classification_report(y_test,lg_pred_tf))
```
